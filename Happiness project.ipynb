# Remove Outlier

```python
# Remove outlier from family
```

```python
df['Family'].describe()
```

```python
# assuming that family follow Guessuion distribution we will calculated boundaries which are differentiate the outlier

upper_boundary=df['Family'].mean() + 3 * df['Family'].std()
lower_boundary=df['Family'].mean() - 3 * df['Family'].std()

print(upper_boundary)
print(lower_boundary)
print(df['Family'].mean())
```

```python
#lets compute inter-qurtiel range

IQR=df.Family.quantile(0.75)-df.Family.quantile(0.25)
IQR
```

```python
lower_bridge=df['Family'].quantile(0.25)-(IQR*1.5)
upper_bridge=df['Family'].quantile(0.75)+(IQR*1.5)
print(lower_bridge)
print(upper_bridge)
```

```python
# Extream outlier

lower_bridge=df['Family'].quantile(0.25)-(IQR*3)
upper_bridge=df['Family'].quantile(0.75)+(IQR*3)
print(lower_bridge)
print(upper_bridge)
```
